```python
# Import dependices
import pandas as pd
import requests
import json
import csv
```

```python
# Bring in player_ids.csv
file = "player_list.csv"
player_list_df = pd.read_csv(file)
player_list_df.head()
```

```python
# Loop through player_ids
player_ids = player_list_df['player_id'].tolist()
player_ids
```

```python
# Create an empty list where shot data will be stored
shot_list = []
```

```python
## I wanted to create a function that we can use for the shortchart NBA Stat APIs
## We can use this logic to loop over the shortchartdetail by updating the URL and paramters

## What I am thinking to create another function getShots(player_id)

# My reqest kept timing out so I reffered to a stackoverflow article and inserted a header
headers = {
    'user-agent': ('Mozilla/5.0 (Windows NT 6.2; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/57.0.2987.133 Safari/537.36'),
    'Dnt': ('1'),
    'Accept-Encoding': ('gzip, deflate, sdch'),
    'Accept-Language': ('en'),
    'origin': ('http://stats.nba.com')
    }

# ## This function will iterate through each seasons in our list by using a custom NBA Stats API call
def getShots(person):

    # create f string url with a varaible to loop through all player_ids
    url = f'https://stats.nba.com/stats/shotchartdetail?AheadBehind=&ClutchTime=&ContextFilter=&ContextMeasure=PTS&DateFrom=&DateTo=&EndPeriod=&EndRange=&GameID=&GameSegment=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&Period=0&PlayerID={player}&PlayerPosition=&PointDiff=&Position=&RangeType=&RookieYear=&Season=&SeasonSegment=&SeasonType=Regular+Season&StartPeriod=&StartRange=&TeamID=0&VsConference=&VsDivision='
    # create the baseurl with parmaters; added 'timeout' option so if the data is not being scraper it will throw an error
    response = requests.get(url, headers=headers, timeout=60)

    # put the result of the call into player_data to loop through each iteration
    shot_json = response.json()
    
    for shot in shot_json["resultSets"][0]["rowSet"]:
        shotDict = {
            "Player_ID": shot[3],
            "Player_Name": shot[4],
            "Team_ID": shot[5],
            "Team_Name": shot[6],
            "Loc_X": shot[17],
            "Loc_Y": shot[18],
            "Game_Date": shot[21]
        }
        shot_list.append(shotDict)
```

```python
# run our function to return values for each season in our 'seasons' list
for player in player_ids:
    # put this print statement to make sure the data was being scrapped
    print(player)
    getShots(player)
```

```python
shot_list_df = pd.DataFrame(shot_list)
shot_list_df.head()
```
