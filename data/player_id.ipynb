```python
# Import dependices
import pandas as pd
import requests
import json
import csv
```

```python
# Create an empty list where 'Player' and 'player_id' will be stored
player_list = []

# Loop over the each seasons since 2003
seasons = ["2003-04", "2004-05", "2005-06", "2006-07", "2007-08", 
         "2008-09", "2009-10", "2010-11", "2011-12", "2012-13", "2013-14", 
         "2014-15", "2015-16", "2016-17", "2017-18", "2018-19"]
```

```python
## I wanted to create a function that we can use for the shortchart NBA Stat APIs
## We can use this logic to loop over the shortchartdetail by updating the URL and paramters

# My reqest kept timing out so I reffered to a stackoverflow article and inserted a header
headers = {
    'user-agent': ('Mozilla/5.0 (Windows NT 6.2; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/57.0.2987.133 Safari/537.36'),
    'Dnt': ('1'),
    'Accept-Encoding': ('gzip, deflate, sdch'),
    'Accept-Language': ('en'),
    'origin': ('http://stats.nba.com')
}


## This function will iterate through each seasons in our list by using a custom NBA Stats API call
def getPlayers(year):

    paramters = {
        "Season": season,
        "LeagueID": "00",
        "IsOnlyCurrentSeason": "1"
    }
    
    # create the baseurl with parmaters; added 'timeout' option so if the data is not being scraper it will throw an error
    response = requests.get("http://stats.nba.com/stats/commonallplayers", params=paramters, headers=headers, timeout=60)

    # put the result of the call into player_data to loop through each iteration
    player_data = response.json()

    # this return each player with player_id
    for player in player_data["resultSets"][0]["rowSet"]:
        playerDict = {
            "player": player[2],
            "player_id": player[0]
        }
        
        # checks for duplicates
        if not any(d['player'] == player[2] for d in player_list):
            player_list.append(playerDict)
```

```python
# run our function to return values for each season in our 'seasons' list
for season in seasons:
    # put this print statement to make sure the data was being scrapped
    print(season)
    getPlayers(season)
```
