```python
shotlist2_df = newdf
shotlist2_df.head()
```

```python
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20031018) & (shotlist2_df.Game_Date < 20040530), 'Season'] = '2003-04'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20041018) & (shotlist2_df.Game_Date < 20050530), 'Season'] = '2004-05'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20051018) & (shotlist2_df.Game_Date < 20060530), 'Season'] = '2005-06'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20061018) & (shotlist2_df.Game_Date < 20070530), 'Season'] = '2006-07'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20071018) & (shotlist2_df.Game_Date < 20080530), 'Season'] = '2007-08'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20081018) & (shotlist2_df.Game_Date < 20090530), 'Season'] = '2008-09'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20091018) & (shotlist2_df.Game_Date < 20100530), 'Season'] = '2009-10'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20101018) & (shotlist2_df.Game_Date < 20110530), 'Season'] = '2010-11'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20111018) & (shotlist2_df.Game_Date < 20120530), 'Season'] = '2011-12'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20121018) & (shotlist2_df.Game_Date < 20130530), 'Season'] = '2012-13'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20131018) & (shotlist2_df.Game_Date < 20140530), 'Season'] = '2013-14'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20141018) & (shotlist2_df.Game_Date < 20150530), 'Season'] = '2014-15'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20151018) & (shotlist2_df.Game_Date < 20160530), 'Season'] = '2015-16'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20161018) & (shotlist2_df.Game_Date < 20170530), 'Season'] = '2016-17'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20171018) & (shotlist2_df.Game_Date < 20180530), 'Season'] = '2017-18'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20181018) & (shotlist2_df.Game_Date < 20190530), 'Season'] = '2018-19'
```

```python
shotlist2_df
```

```python
shotlist3_df = shotlist2_df.dropna(subset=['Season'])
shotlist3_df
```

```python
shotlist3_df.reset_index(drop=True)
```
